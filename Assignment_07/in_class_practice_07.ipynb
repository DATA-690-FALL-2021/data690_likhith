```python
import pandas as pd

print("pandas version is", pd.__version__)

import plotly
print("plotly version is", plotly.__version__)

import plotly.graph_objects as go

import plotly.express as px
```

```python
df = pd.read_csv("/content/wdi_data.csv")
df.head()
```

```python
#df.drop(columns=["Unnamed: 0"], inplace=true)

df_new = df.drop(columns=["Unnamed: 0"])
df_new.head()
```

```python
# df_ind = df_new.query('Country Code=="IND"')
# did not work as there is a space in beterrn country and code in the coloumn name  
```

```python
df_ind = df_new[df_new["Country Code"]== "IND"]
df_ind
```

```python
df_ind_gdppcap = df_ind.query('indicator =="NY.GDP.PCAP.PP.CD"')
df_ind_gdppcap
```

```python
fig = go.Figure()

trace_0 = go.Bar(
    x=df_ind_gdppcap["Year"],
    y=df_ind_gdppcap["value"]
)

fig.add_trace(trace_0)

fig.update_layout(
    title="INDIA GDP per capita Over Time",
    xaxis_title="Year",
    yaxis_title="GDP per Capita"
)

fig.show()
```

```python
fig = px.bar(
    data_frame=df_ind_gdppcap,
    title="INDIA GDP per capita Over Time",
    labels={"value":"GDP per Capita","Year":"YEAR"},
    x="Year",
    y="value"
)

fig.show()
```

# making the grapf horzontal 

```python
fig = px.bar(
    data_frame=df_ind_gdppcap,
    title="INDIA GDP per capita Over Time",
    labels={"value":"GDP per Capita","Year":"YEAR"},
    orientation="h",
    y="Year",
    x="value"
)

fig.update_xaxes(tickangle=45)

fig.show()
```
