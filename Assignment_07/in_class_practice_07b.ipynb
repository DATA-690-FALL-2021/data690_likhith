```python
import pandas as pd

print("pandas version is", pd.__version__)

import plotly
print("plotly version is", plotly.__version__)

import plotly.graph_objects as go

import plotly.express as px
```

```python
df = pd.read_csv("/content/wdi_data.csv")
df.head()
```

```python
#df.drop(columns=["Unnamed: 0"], inplace=true)

df_new = df.drop(columns=["Unnamed: 0"])
df_new.head()
```

```python
# df_ind = df_new.query('Country Code=="IND"')
# did not work as there is a space in beterrn country and code in the coloumn name  
```

```python
df_ind_bra = df_new[df_new["Country Code"].isin(['IND','BRA'])]
df_ind_bra.head()
```

```python
df_gdp = df_ind_bra.query('indicator =="NY.GDP.PCAP.PP.CD"')
df_gdp
```

```python
df_ind = df_gdp[df_gdp["Country Code"]=="IND"]
df_ind
```

```python
df_bra = df_gdp[df_gdp["Country Code"]=="BRA"]
df_bra
```

```python
fig = go.Figure()

trace_ind = go.Bar(
    x=df_ind["Year"],
    y=df_ind["value"],
    name="INDIA"
)
fig.add_trace(trace_ind)

trace_bra = go.Bar(
    x=df_bra["Year"],
    y=df_bra["value"],
    name="BRAZIL"
)
fig.add_trace(trace_bra)

fig.update_layout(
    title="INDIA and BRAZIL GDP per capita Over Time",
    xaxis_title="Year",
    yaxis_title="GDP per Capita"
)

fig.show()
```

```python
df_ind_bra = df_gdp[df_gdp["Country Code"].isin(["IND","BRA"])]
df_ind_bra
```

# using plotly express
- barmode defines the way bar graph comes,
- color is used to differentiate 

```python
fig = px.bar(
    data_frame=df_ind_bra,
    title="INDIA GDP per capita Over Time",
    labels={"value":"GDP per Capita","Year":"YEAR"},
    color= 'Country Code',
    barmode="group",
    x="Year",
    y="value"
)

fig.show()
```

## Stacked Bar chart 

```python
fig = px.bar(
    data_frame=df_ind_bra,
    title="INDIA GDP per capita Over Time",
    labels={"value":"GDP per Capita","Year":"YEAR"},
    color= 'Country Code',
    x="Year",
    y="value"
)

fig.show()
```
