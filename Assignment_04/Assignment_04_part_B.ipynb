## Cleaning of the year coloumn
- defining the cleanse_year function by replacing "*" in the coloumn

```python
def cleanse_year(x):
    return int(x.replace("*",""))
```

- appending the above definition to our file 

```python
df["clean_year_int"] = df["Census Year"].apply(cleanse_year)
df
```

## Cleaning the population colomn by removing the " , "

```python
def cleanse_pop(a_string):
    return int(a_string.replace(",",""))
```

- applying the same on to the fuke to create a int population count

```python
df["total_population_int"] = df["Total Population"].apply(cleanse_pop)
```

```python
df
```

## Cleaning the Census Cost coloumn by replacing the $ and billion in it
- defining the function for the same 

```python
def cleanse_cost(b_string):
    return (b_string.replace(",","").replace("$","").replace(" Billion","000000000"))
  
```

-applying the definitaion from above to the file to remove the same and make it into an int

```python
df["census_cost_int"] = df["Census Cost"].apply(cleanse_cost)
df['census_cost_int'] = pd.to_numeric(df['census_cost_int'])
df
```

## cleaning the Average Cost Per Person coloumn from the file
- removing the cents and $ definition and applying the same.

```python
def cleanse_Average(c_string):
    return(c_string.replace(",","").replace("$","").replace("cents",""))
```

```python
df["Avg_Cost_Per_Person_INT"] = df["Average Cost Per Person"].apply(cleanse_Average)
df
```
