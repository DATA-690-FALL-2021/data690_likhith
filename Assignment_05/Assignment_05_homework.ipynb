## Filter the smaller dataset further to only keep colleges from Maryland, Virgina, Washington DC, Delaware, and Pennsylvinia.
- using "isin" function from smaller dataset to combine the specific state name and get the list.
- https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.isin.html

```python
Five_States_Colleges = Smaller_data[Smaller_data['CITY'].isin(['Delaware', 'Washington','Virginia','Baltimore','Pennsylvania'])]
Five_States_Colleges = pd.DataFrame(Five_States_Colleges)
Five_States_Colleges
```
